La multicolin√©arit√© va biaiser nos r√©sidus. Que faire ?


Le VIF (Variance Inflation Factor) est un indicateur plus pr√©cis que la simple corr√©lation de Pearson pour d√©tecter la multicolin√©arit√© entre les variables explicatives. Alors que la corr√©lation de Pearson mesure uniquement la relation entre deux variables √† la fois, le VIF prend en compte l‚Äôensemble des relations lin√©aires entre une variable et toutes les autres, ce qui permet de mieux √©valuer sa redondance r√©elle dans le mod√®le.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

X = df[['PM2_5', 'SO2', 'NO2', 'O3', 'CO']]  # tes variables
vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif)
```

Apr√®s calcul des VIFs, la variable CO va √™tre retir√©e en raison d‚Äôun VIF sup√©rieur √† 10, indiquant une multicolin√©arit√© s√©v√®re.

## 3. Les facteurs explicatifs de la qualit√© de l‚Äôair

### 3.1 Mod√©lisation et v√©rification des hypoth√®ses

#### 3.1.1 Mod√©lisation

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm


train_df = X_train.copy()
train_df['AQI'] = y_train.values  # on ajoute la cible

# Variables num√©riques
base_vars = ['PM2_5','NO2', 'SO2', 'O3', 'Hour', 'Month']

# Variables de ville (d√©j√† en dummies dans ton X_train)
city_vars = [c for c in train_df.columns if c.startswith('City_')]
city_terms = ' + '.join(city_vars)

# Formule simple (sans interaction)
formula = 'AQI ~ ' + ' + '.join(base_vars) + ' + ' + city_terms
print("Formule du mod√®le :", formula)

model = smf.ols(formula=formula, data=train_df)
model = model.fit()
print(model.summary())
```

### 3.1.2 V√©rification des hypoth√®ses

#### 1. Lin√©arit√© de la relation (AQI ‚Üî variables)

La condition 3 est v√©rifi√©e alors la relation est lin√©aire.

Hypoth√®se v√©rifi√©e ! 

#### 2. Normalit√© des r√©sidus

```python
def plot_residuals_density(residuals: df, colors: list):
    plt.figure(figsize=(12, 8))
    sns.kdeplot(residuals, fill=True, color=colors[1], alpha=0.6)
    plt.axvline(
        np.mean(residuals),
        color=colors[0],
        linestyle="--",
        label="Moyenne des r√©sidus",
    )
    p = plt.title("Distribution des r√©sidus", fontweight="bold")
    plt.grid(True)
    sns.kdeplot(
        np.random.normal(np.mean(residuals), np.std(residuals), 10000),
        color=colors[2],
        linestyle="--",
        label="Distribution normale",
    )
    plt.legend()
    plt.show()
```

```python
plot_residuals_density(model.resid, ['red', 'blue', 'green'])
```

Nos r√©sidus ne sont pas gaussiens mais la moyenne des r√©sidus est bien centr√©e en 0. Nous avons donc v√©rif√© l'hypoth√®se de normalit√© des r√©sidus.

Hypoth√®se v√©rifi√©e ! 

#### 3. Homoscedasticit√© (variance constante)

```python
def plot_homoscedasticity(residuals, preds, colors, exog):
    import statsmodels.stats.api as sms
    from statsmodels.compat import lzip

    name = ["F statistic", "p-value"]
    test = sms.het_goldfeldquandt(residuals, exog)
    stat, p = lzip(name, test)[1]

    plt.figure(figsize=(12, 8))
    plt.scatter(preds, residuals, color=colors[1], alpha=0.8)
    plt.xlabel("Valeurs pr√©dites par le mod√®le")
    plt.ylabel("R√©sidus")
    plt.axhline(0, color=colors[0], linestyle="--")
    plt.title(
        f"R√©sidus vs. valeurs pr√©dites - Gold-Felquant p-value: {p:.2f}",
        fontweight="bold",
    )
    plt.show()
```

```python
plot_homoscedasticity(model.resid, model.fittedvalues, ['red', 'blue', 'green'], exog=model.model.exog)
```

La p-value est sup√©rieure √† 0.05, donc l'hypoth√®se d'homoscedasticit√© est v√©rifi√©e selon le test de Gold-Felquant.

Hypoth√®se v√©rifi√©e ! 

#### 4. Ind√©pendance des erreurs (auto-corr√©lation)

```python
from statsmodels.stats.stattools import durbin_watson

dw = durbin_watson(model.resid)
print("Durbin-Watson =", round(dw,2))
```

Si Durbin-Watson = 2, cela signifie qu‚Äôil n‚Äôy a pas d‚Äôautocorr√©lation des r√©sidus dans notre mod√®le.

Hypoth√®se v√©rifi√©e ! 

Nous avons le BLUE (Best Linear Unbiased Estimator) !

## 4. Les facteurs sont-ils les m√™mes en fonction des villes ?

Les r√©sultats montrent que les facteurs explicatifs de l‚ÄôAQI varient fortement d‚Äôune ville √† l‚Äôautre. Les interactions significatives entre les polluants et les villes indiquent que l‚Äôimpact d‚Äôun m√™me polluant sur l‚ÄôAQI d√©pend du contexte local.
Par exemple, l‚Äôeffet de PM2.5 est bien plus marqu√© √† Duba√Ø qu‚Äô√† Sydney, et l‚Äôozone (O3) a un effet oppos√© selon les villes.


Afin d'√©viter les probl√®mes de multicolin√©arit√© li√©s aux variables indicatrices et aux interactions, nous avons choisi d'estimer un mod√®le s√©par√© pour chaque ville. Cette approche permet d'identifier plus clairement les effets sp√©cifiques des polluants dans chaque contexte local, sans que les effets crois√©s ne viennent biaiser les r√©sultats. Les mod√®les obtenus sont directement comparables, et les r√©sultats confirment que les facteurs explicatifs de l‚ÄôAQI varient d'une ville √† l'autre.

```python
import statsmodels.formula.api as smf
import pandas as pd

# Variables explicatives
base_vars = ['PM2_5', 'NO2', 'SO2', 'O3', 'Hour', 'Month']
formule_base = 'AQI ~ ' + ' + '.join(base_vars)

# Liste des villes (colonnes de dummies)
villes = ['City_Cairo', 'City_Dubai', 'City_London', 'City_New_York', 'City_Sydney']

# Copie du DataFrame d'entra√Ænement
train_df = X_train.copy()
train_df['AQI'] = y_train.values

# Dictionnaire pour stocker les mod√®les
models_par_ville = {}

for ville_dummy in villes:
    ville_nom = ville_dummy.replace("City_", "")
    
    print(f"\nüìç Mod√®le pour la ville : {ville_nom}")
    
    # Filtrer les lignes o√π la ville est active
    df_ville = train_df[train_df[ville_dummy] == 1].copy()
    
    # V√©rifier qu'on a assez d'observations
    if df_ville.shape[0] < 30:
        print(f"‚ö†Ô∏è Trop peu d'observations pour {ville_nom}. Mod√®le ignor√©.")
        continue
    
    # Estimer le mod√®le
    model = smf.ols(formula=formule_base, data=df_ville).fit()
    models_par_ville[ville_nom] = model
    
    # Affichage du r√©sum√©
    print(model.summary())
    print("=" * 100)
```

üîç Pourquoi les R¬≤ sont plus faibles maintenant ?

- Moins de donn√©es dans chaque mod√®le
- Quand nous faisons un mod√®le global, nous exploitons toutes les donn√©es en m√™me temps :
    ‚Üí √áa peut donner un R¬≤ globalement √©lev√©, m√™me si les relations varient selon les villes.

- Mais maintenant que nous s√©parons les villes, chaque mod√®le a :

    - moins d'observations,

    - moins de variabilit√© globale √† expliquer,

    - une structure plus locale (plus r√©aliste, mais moins "optimis√©e").

Il est alors normal que le R¬≤ descende un peu, m√™me si le mod√®le est bon.

```python
import pandas as pd

# Variables √† comparer
polluants = ['PM2_5', 'NO2', 'SO2', 'O3']

# Initialisation du tableau
comparaison = pd.DataFrame(columns=polluants)

# Remplir le tableau avec les coefficients de chaque ville
for ville, model in models_par_ville.items():
    coeffs = model.params
    ville_nom = ville.replace("City_", "")
    comparaison.loc[ville_nom] = [coeffs.get(var, float('nan')) for var in polluants]

# Afficher les r√©sultats
print("üßæ Comparaison des effets des polluants sur l‚ÄôAQI selon les villes :")
display(comparaison.round(4))
```

```python
import statsmodels.formula.api as smf

train_df = X_train.copy()
train_df['AQI'] = y_train.values  # on ajoute la cible

# Variables num√©riques
base_vars = ['PM2_5','NO2', 'SO2','O3', 'Hour']

# Variables de ville (en dummies)
city_vars = [c for c in train_df.columns if c.startswith('City_')]
city_terms = ' + '.join(city_vars)

# Boucle sur les mois
for month in sorted(train_df['Month'].unique()):
    df_mois = train_df[train_df['Month'] == month].copy()

    formula = 'AQI ~ ' + ' + '.join(base_vars) + ' + ' + city_terms
    print(f"\nüìÖ Mod√®le pour le mois : {month}")
    print(f"Formule utilis√©e : {formula}")

    model = smf.ols(formula=formula, data=df_mois).fit()
    print(model.summary())
```

```python
import pandas as pd
import statsmodels.formula.api as smf

# Initialisation
polluants = ['PM2_5', 'NO2', 'SO2', 'O3']
coeffs_par_mois = pd.DataFrame(columns=polluants)

# Pr√©paration des donn√©es
train_df = X_train.copy()
train_df['AQI'] = y_train.values

# Variables du mod√®le
base_vars = polluants + ['Hour']
city_vars = [c for c in train_df.columns if c.startswith('City_')]
city_terms = ' + '.join(city_vars)

# Boucle sur chaque mois
for month in sorted(train_df['Month'].unique()):
    df_mois = train_df[train_df['Month'] == month].copy()
    formula = 'AQI ~ ' + ' + '.join(base_vars) + ' + ' + city_terms
    model = smf.ols(formula=formula, data=df_mois).fit()
    
    # Extraire les coefficients uniquement des polluants
    coeffs = model.params[polluants]
    coeffs_par_mois.loc[month] = coeffs

# Affichage final du DataFrame
print("\nüìä Coefficients des polluants par mois :\n")
coeffs_par_mois.round(3)
```

```python
import matplotlib.pyplot as plt

# üî∏ Liste des polluants
polluants = ['PM2_5', 'NO2', 'SO2', 'O3']

# üî∏ Palette de couleurs personnalis√©e (facultatif mais clair)
colors = {
    'PM2_5': 'orange',
    'NO2': 'red',
    'SO2': 'deeppink',
    'O3': 'magenta'
}

# üî∏ Nommer l'index pour clart√©
coeffs_par_mois.index.name = 'Mois'

# üî∏ Cr√©ation du graphique
plt.figure(figsize=(14, 6))

for pollutant in polluants:
    plt.plot(coeffs_par_mois.index, coeffs_par_mois[pollutant],
             marker='o', label=pollutant, color=colors.get(pollutant))

plt.title("üìà √âvolution mensuelle des coefficients des polluants (effet sur l‚ÄôAQI)")
plt.xlabel("Mois")
plt.ylabel("Coefficient")
plt.xticks(range(1, 13))
plt.grid(True)
plt.legend(title="Polluants")
plt.tight_layout()
plt.show()
```

Cette approche permet d'identifier plus clairement les effets sp√©cifiques des polluants pour chaque heure, sans que les effets crois√©s ne viennent biaiser les r√©sultats.

#### Par heure

```python
import statsmodels.formula.api as smf

# Cr√©ation de la base d'entra√Ænement compl√®te
train_df = X_train.copy()
train_df['AQI'] = y_train.values  # on ajoute la cible

# Variables explicatives
base_vars = ['PM2_5', 'NO2', 'SO2', 'O3', 'Month']  # cette fois on garde Month, pas Hour
city_vars = [c for c in train_df.columns if c.startswith('City_')]
city_terms = ' + '.join(city_vars)

# Boucle sur les heures (0 √† 23)
for hour in sorted(train_df['Hour'].unique()):
    df_heure = train_df[train_df['Hour'] == hour].copy()

    # Formule du mod√®le
    formula = 'AQI ~ ' + ' + '.join(base_vars) + ' + ' + city_terms
    print(f"\nüïí Mod√®le pour l'heure : {hour}")
    print(f"Formule utilis√©e : {formula}")

    model = smf.ols(formula=formula, data=df_heure).fit()
    print(model.summary())
```

```python
import pandas as pd
import statsmodels.formula.api as smf

# D√©finir les variables polluantes
polluants = ['PM2_5', 'NO2', 'SO2', 'O3']
coeffs_par_heure = pd.DataFrame(columns=polluants)

# Pr√©parer les donn√©es
train_df = X_train.copy()
train_df['AQI'] = y_train.values

# Variables du mod√®le
base_vars = polluants + ['Month']  # on garde Month au lieu de Hour ici
city_vars = [c for c in train_df.columns if c.startswith('City_')]
city_terms = ' + '.join(city_vars)

# Boucle sur chaque heure de la journ√©e
for hour in sorted(train_df['Hour'].unique()):
    df_heure = train_df[train_df['Hour'] == hour].copy()
    formula = 'AQI ~ ' + ' + '.join(base_vars) + ' + ' + city_terms
    model = smf.ols(formula=formula, data=df_heure).fit()

    # Extraire les coefficients uniquement des polluants
    coeffs = model.params[polluants]
    coeffs_par_heure.loc[hour] = coeffs

# Affichage du tableau final
print("\nüìä Coefficients des polluants par heure :\n")
coeffs_par_heure.round(3)
```

```python
import matplotlib.pyplot as plt

# Taille de la figure
plt.figure(figsize=(12, 6))

# Tracer chaque polluant
for col in coeffs_par_heure.columns:
    plt.plot(coeffs_par_heure.index, coeffs_par_heure[col], marker='o', label=col)

# Personnalisation du graphique
plt.title("√âvolution des coefficients des polluants en fonction de l'heure")
plt.xlabel("Heure de la journ√©e")
plt.ylabel("Coefficient dans le mod√®le AQI")
plt.legend(title="Polluants")
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()

# Affichage
plt.show()
```

Les facteurs (polluants) ne sont pas constants dans le temps.
Leurs coefficients varient significativement d‚Äôun mois et d'une heure √† l‚Äôautre, notamment pour PM2.5 et NO2, ce qui indique que leur effet sur l‚ÄôAQI d√©pend du mois.

### 6. Ouverture

#### Analyse des pics de pollution via une mod√©lisation binaire
Objectif : 

L‚Äôobjectif de cette analyse est de mod√©liser la probabilit√© d‚Äôun √©pisode de pollution √©lev√©e (d√©fini ici comme des cas o√π l‚ÄôAQI d√©passe 100, un niveau "non sain") √† partir des principaux polluants atmosph√©riques et du mois de l‚Äôann√©e. Cela permet de compl√©ter l‚Äôanalyse lin√©aire par une approche binaire orient√©e ‚Äúrisque‚Äù.

Cela revient √† r√©pondre √† une question oui/non :
-> ‚ÄúEst-ce qu‚Äôon entre dans une situation dangereuse pour la sant√© ?‚Äù


L'int√©r√™t serait donc concr√®tement de pouvoir alerter la population, d√©clencher les mesures d'urgences... pour mieux r√©pondre √† la situation.

```python
# 1. Cr√©er la variable binaire Pollution_Elevee
# Variable binaire : 1 si AQI > 100, 0 sinon
df['Pollution_Elevee'] = (df['AQI'] > 100).astype(int)
print(df['Pollution_Elevee'].value_counts())
```

```python
# Liste des variables num√©riques
vars_num = ['PM2_5', 'NO2', 'SO2', 'O3', 'Month']

# Taille de la figure globale
plt.figure(figsize=(15, 12))

# Boucle sur chaque variable
for i, var in enumerate(vars_num, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(x='Pollution_Elevee', y=var, data=df)
    plt.title(f'{var} selon Pollution √âlev√©e')
    plt.xlabel('Pollution √âlev√©e (0 = Non, 1 = Oui)')
    plt.ylabel(var)

plt.tight_layout()
plt.show()
```

```python
# Formule du mod√®le
logit_formula = 'Pollution_Elevee ~ PM2_5 + NO2 + SO2 + O3 + Month'

# Ajustement du mod√®le
logit_model = smf.logit(formula=logit_formula, data=df).fit()

# R√©sum√© du mod√®le
print(logit_model.summary())
```

Tous les coefficients sont statistiquement significatifs √† p < 0.05.
- PM2_5: ‚Üó Augmentation du PM2.5 ‚Ü¶ augmente fortement la probabilit√© de pollution √©lev√©e. C‚Äôest le facteur principal.
- NO2: ‚Üò L‚Äôeffet est n√©gatif, contre-intuitif √† premi√®re vue mais peut s'expliquer par une corr√©lation indirecte (par ex. plus de NO2 dans des mois moins pollu√©s globalement, ou compensation avec O3).
- SO2:‚Üò L√©g√®re contribution n√©gative.
- O3: ‚Üó Contribution mod√©r√©e mais significative dans l‚Äôaugmentation du risque.

```python
# Nous allons utiliser la matrice de confusion pour √©valuer la performance de notre mod√®le

from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# √âtape 1 : pr√©dictions avec le mod√®le logit
y_pred = logit_model.predict(df[['PM2_5', 'NO2', 'SO2', 'O3', 'Month']])
y_pred_class = (y_pred > 0.5).astype(int)  # seuil de 0.5

# √âtape 2 : matrice de confusion
cm = confusion_matrix(df['Pollution_Elevee'], y_pred_class)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non Pollu√©", "Pollution √âlev√©e"])
disp.plot(cmap='Blues')
plt.title("Matrice de Confusion")
plt.show()

# √âtape 3 : rapport de classification
print(classification_report(df['Pollution_Elevee'], y_pred_class, target_names=["Non Pollu√©", "Pollution √âlev√©e"]))
```
